```python
import numpy as np
```

```python
class BackPropagation(object):
    def __init__(self):
        self.no_input = 3
        self.no_hidden = 2
        self.no_output = 2
        
        self.X = np.array([1, 0, 1], dtype = float)
        self.T = np.array(([1, 0]), dtype = float)
        
        "self.W_ih = np.ndarray(shape=(self.no_hidden, self.no_input), dtype=float)"
        self.W_ih = np.array(([0.2, 0.4, -0.5],
                             [-0.3, 0.1, 0.2]), dtype = float)
        "self.W_oh = np.ndarray(shape=(self.no_output, self.no_hidden), dtype=float)"
        self.W_oh = np.array(([-0.3, -0.2],
                             [0.4, 0.6]), dtype = float)
        
        self.hidden_bias = np.array([-0.4, 0.2], dtype = float)
        self.output_bias = np.array([0.1, 0.2], dtype = float)
        
        self.I_hidden = np.zeros(self.no_hidden, dtype = float)
        self.O_hidden = np.zeros(self.no_hidden, dtype = float)
        
        self.I_output = np.zeros(self.no_output, dtype = float)
        self.O_output = np.zeros(self.no_output, dtype = float)
        
        self.Err_output = np.zeros(self.no_output, dtype = float)
        self.Err_hidden = np.zeros(self.no_hidden, dtype = float)
        
    def sigmoid(self, Ij):
        return (1/(1+(np.exp(-Ij))))
    
    def backpropagation(self):
        for epoch in range(15):
            print("Epoch-" , epoch)
            print("---------------------")
            "for each hidden layer calculate the input and output to the hidden layer"
            for i in range(self.no_hidden):
                for j in range(self.no_input):
                    self.I_hidden[i] = self.I_hidden[i] + (self.W_ih[i, j] * self.X[j])
                self.I_hidden[i] = self.I_hidden[i] + self.hidden_bias[i]
                self.O_hidden[i] = self.sigmoid(self.I_hidden[i])

            print("Hidden Layer(Input)" , self.I_hidden)
            print("")
            print("Hidden Layer(Output)", self.O_hidden)
            print("")
            
            "for each output layer calculate the input and output"
            for i in range(self.no_output):
                for j in range(self.no_hidden):
                    self.I_output[i] = self.I_output[i] + (self.W_oh[i, j] * self.O_hidden[j])
                self.I_output[i] = self.I_output[i] + self.output_bias[i]
                self.O_output[i] = self.sigmoid(self.I_output[i])

            print("Output Layer(Input)" , self.I_output)
            print("")
            print("Output Layer(Output)", self.O_output)
            print("")

            "for each output layer calculate the Error"
            for i in range(self.no_output):
                self.Err_output[i] = self.O_output[i] * (1 - self.O_output[i]) * (self.T[i] - self.O_output[i])
            print("Err-Output:" , self.Err_output)

            "for each hidden layer calculate the Error"
            for i in range(self.no_hidden):
                sum = 0.0
                for j in range(self.no_output):
                    sum = sum + (self.Err_output[j] * self.W_oh[j, i])
                self.Err_hidden[i] = self.O_hidden[i] * (1 - self.O_hidden[i]) * sum
            print("Err-hidden" , self.Err_hidden)
            print("")

            "update weights associated to hidden to output layer"
            for i in range(self.no_hidden):
                for j in range(self.no_output):
                    self.W_oh[j, i] = self.W_oh[j, i] + (self.Err_output[j] * self.O_hidden[i])
            
            print("Updated Weight(From Hidden To Output Layer): ")
            print(self.W_oh)
            print("")

            "update weights associated with the hidden layer"
            for i in range(self.no_hidden):
                self.hidden_bias[i] = self.hidden_bias[i] + self.Err_hidden[i]

            print("Updated bias at hidden layer:")
            print(self.hidden_bias)
            print("")

            "update weights associated to input to hidden layer"
            for i in range(self.no_hidden):
                for j in range(self.no_input):
                    self.W_ih[i, j] = self.W_ih[i, j] + (self.Err_hidden[i] * self.X[j])
            print("Updated Weights (From Input to Hidden layer):")
            print(self.W_ih)
            print("")

            "update biases at output layer"
            for i in range(self.no_output):
                    self.output_bias[i] = self.output_bias[i] * self.Err_output[i]
            print("Updated bias at the output layer:")
            print(self.output_bias)
            print("")

bp = BackPropagation()
bp.backpropagation()
        
```
